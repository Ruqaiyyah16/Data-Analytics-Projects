# 5. MACHINE LEARNING ALGORITHM

LINEAR REGRESSION

```r
library(caTools)
```

total_cases_confirmed

```r
cor(dataset1$total_cases_confirmed,dataset1$total_deaths) 
```

```r
data1 = dataset1
creating_sample1 = sample.split(data1,SplitRatio = 0.7)
training_set1 = subset(data1,split = TRUE)
testing_set1 = subset(data1,split = FALSE)
model1 = lm(data1$total_deaths~data1$total_cases_confirmed,data = data1)
model1
```

```r
predicting1 = predict(model1,testing_set1)
print(predicting1)
```

```r
accuracy1 = sqrt(mean((testing_set1$total_deaths-predicting1)^2))
accuracy1
```

```r
summary(model1)
```
